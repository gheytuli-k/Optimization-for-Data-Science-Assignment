```python

files = []
folder_path = "Instance Files"
for file in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path, file)):
        if file != "A.Readme.txt":
            files.append(os.path.join(folder_path, file))
files
```

```python
def read(file_path):
    with open(file_path, 'r') as file:
        # Step 1: Read the header
        header = file.readline().strip().split()[2]
        num_pairs = int(header)
        
        header = file.readline().strip().split()[2]
        num_ndds = int(header)
        
        header = file.readline().strip().split()[2]
        num_arcs = int(header)

        # Step 2: Read the pairs and NDDs
        pairs_ndds = []
        for _ in range(num_pairs + num_ndds):
            line = file.readline().strip().split(',')
            pair_id = int(line[0])
            is_ndd = int(line[1])  # 0 for pair, 1 for NDD
            donor_blood_type = int(line[2])  # 0 for O, 1 for A, 2 for B, 3 for AB
            patient_blood_type = int(line[3])  # 0 for O, 1 for A, 2 for B, 3 for AB
            patient_vpra = int(line[4])  # 0 for <0.5, 1 for 0.5-0.85, 2 for >0.85

            # Store each pair/NDD as a dictionary
            pairs_ndds.append({
                'pair_id': pair_id,
                'is_ndd': is_ndd,
                'donor_blood_type': donor_blood_type,
                'patient_blood_type': patient_blood_type,
                'patient_vpra': patient_vpra
            })

        # Step 3: Read the arcs
        arcs = []
        for _ in range(num_arcs):
            arc_line = file.readline().strip().split(',')
            donor_pair_id = int(arc_line[0][1:])  # Strip the opening parenthesis
            patient_pair_id = int(arc_line[1].split(')')[0])  # Strip the closing parenthesis
            weight = int(arc_line[2])

            arcs.append({
                'donor_pair_id': donor_pair_id,
                'patient_pair_id': patient_pair_id,
                'weight': weight
            })

    # Step 4: Return the parsed data
    return {
        'num_pairs': num_pairs,
        'num_ndds': num_ndds,
        'pairs_ndds': pairs_ndds,
        'arcs': arcs
    }
    
read(files[0])
    
```
