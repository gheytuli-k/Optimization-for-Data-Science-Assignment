```python
import numpy as np

# Given data
failure_times = np.array([48, 86, 118, 127, 129, 131, 132, 134, 135, 151])
censored_time = 151  # All censored observations are censored at this time
num_censored = 10

# Parameter pairs to evaluate
parameter_pairs = [
    (300, 1),
    (200, 1),
    (250, 2),
    (300, 5),
    (250, 3),
]

# Initialize a list to store results
results = []

# Compute log-likelihood for each parameter pair
for alpha, beta in parameter_pairs:
    # Uncensored log-likelihood components
    ln_f = np.log(beta / alpha) + (beta - 1) * np.log(failure_times / alpha) - (failure_times / alpha) ** beta
    sum_ln_f = ln_f.sum()
    
    # Censored log-likelihood components
    ln_S = - (censored_time / alpha) ** beta
    sum_ln_S = num_censored * ln_S
    
    # Total log-likelihood
    log_likelihood = sum_ln_f + sum_ln_S
    
    # Store the result
    results.append({
        'alpha': alpha,
        'beta': beta,
        'log_likelihood': log_likelihood
    })

# Sort the results by log-likelihood in descending order (most likely to least likely)
results.sort(key=lambda x: x['log_likelihood'], reverse=True)

# Output the ordered parameter pairs
for idx, res in enumerate(results):
    print(f"Rank {idx+1}: (alpha, beta) = ({res['alpha']}, {res['beta']}), Log-Likelihood = {res['log_likelihood']:.4f}")
```
